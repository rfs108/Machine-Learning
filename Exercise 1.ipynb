1) Plotting the GDP as a function of the year:


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

Year = [1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
GDP = [1.015, 1.33, 2.29, 3.26, 4.951, 6.759, 9.366, 13.131, 15.599]

fig, ax = plt.subplots()

plot1 = ax.plot(Year, GDP, 'ro--')
ax.set_title('Real US GDP (in trillions)')
ax.set_xlabel('Year')
ax.set_ylabel('Trillions in $')
```

2) Mathematical relationship between Year and GDP from plot:

From the plot obtained above, there is a possible linear relationship between GDP and Year. A best fitting line through the data points can be found by computing a least square regression line. The least square regression line attempts to minimize the vertical distance from the actual data points. Essentially, this best fit line will minimize the variance or the sum of squares of the errors.

Step 1: 

Regression Line with form: $Y = mX + b $ 
where $ Y = GDP, X = Year $

Step 2:

Find mean for Year and GDP data:  $\overline{X} = Mean(X) = \frac{1}{9}\sum X  ,  \overline{Y} = Mean(Y) = \frac{1}{9}\sum Y $

$\overline{X} = 1970, \overline{Y} = 6.411 $

Step 3: 

Find the slope using: $ m = \frac{\sum(X - \overline{X}) (Y - \overline{Y})}   {\sum(X - \overline{X}) ^ 2} = 0.18565 $

Step 4:

Find b, the y-intercept, where $ Y = \overline{Y}$ and $X = \overline{X}$

$ 6.411 = 0.18565 * 1970 + b $  therefore $ b = -359.319 $

Step 5:

Our final regression line: $Y = 0.18565X - 359.319$ 

Step 6: 

Correlation coefficient for the regression line can be computed using:
$ R = \frac{9\sum{xy} - (\sum{x})(\sum{y})} {\sqrt{ [9\sum{x^2} - (\sum{x})^2] [ 9\sum{y^2} - (\sum{y})^2]  }} = 0.96426 $

The correlation coefficient R is close to 1 indicating a strong positive linear relationship between Year and GDP.
